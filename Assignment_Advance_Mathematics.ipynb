```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv", encoding="latin1")
x = df["no2"].dropna().values
```

```python
#given
#   ar= 0.05 * ( r mod 7)
#   br= 0.3 * ( r mod 5 + 1)

r = 102303137

ar = 0.05 * ( r % 7)
br = 0.3 * ((r % 5) + 1)

z = x + ar * np.sin(br * x)
```

```python
mu = np.mean(z)

variance = np.var(z)
lamb = 1 / (2 * variance)

c = np.sqrt(lamb / np.pi)

print("Value of mu   :",mu)
print("Value of lamb :",lamb)
print("Value of c    :",c)
```
